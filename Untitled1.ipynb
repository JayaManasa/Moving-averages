```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import json
%matplotlib inline
```

```python
data = pd.read_json('day.json')
```

```python
data.columns = ['date', 'open', 'high', 'low', 'close', 'volume'] 
print(data.head(6))
```

```python
data['col_total'] = data.iloc[:, 1:5].sum(axis = 1)/4
print(data.head(6))
```

```python
data['MA50'] = data['close'].rolling(51).mean()
data['MA200'] = data['close'].rolling(201).mean()
print(data.head(202))
```

```python
data[['close','MA200', 'MA50']].plot(
  label='Moving average', figsize=(15, 10))
```

```python

BUY = data['MA50']>data['MA200']
SELL = data['MA50']<data['MA200']
conditions = [BUY, SELL]
choices = ['BUY', 'SELL']
data['signal'] =np.select(conditions, choices, default = '0')
pd.set_option('display.max_rows', None)
data
```
